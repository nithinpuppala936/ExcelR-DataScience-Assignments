```python
# Identify categorical columns (assuming any column with object dtype is categorical)
categorical_cols = df.select_dtypes(include='object').columns

if len(categorical_cols) == 0:
    print("No categorical columns found in the DataFrame.")
else:
    for col in categorical_cols:
        plt.figure(figsize=(10, 6))
        df[col].value_counts().plot(kind='bar')
        plt.title(f'Frequency of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```

```python
# Generate scatter plots for a few pairs of variables (you can choose specific pairs or iterate through some)
# Let's select a few columns to demonstrate
selected_cols = ['LB', 'AC', 'FM', 'UC', 'DL']

for i in range(len(selected_cols)):
    for j in range(i + 1, len(selected_cols)):
        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=df, x=selected_cols[i], y=selected_cols[j])
        plt.title(f'Scatter plot of {selected_cols[i]} vs {selected_cols[j]}')
        plt.show()
```

```python
# Generate a correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Variables')
plt.show()
```
