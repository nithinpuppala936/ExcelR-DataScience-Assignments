# Bar Chart Analysis for Categorical Columns

```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_columns = ['Day', 'City', 'BU', 'Brand']

for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=d1, x=col, order=d1[col].value_counts().index)
    plt.title(f'Distribution of Sales by {col}')
    plt.xlabel(col)
    plt.ylabel('Number of Sales')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

# Calculating Z-Score

```python
# Calculate the mean and standard deviation for the selected columns
means = d1[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].mean()
stds = d1[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].std()

# Calculate the z-score
z_scores = (d1[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']] - means) / stds

# Display the z-scores
display(z_scores)
```
