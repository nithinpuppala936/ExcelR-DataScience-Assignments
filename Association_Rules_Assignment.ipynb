```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
# Load Online Retail Dataset
df = pd.read_excel('Online retail.xlsx')  # Place OnlineRetail.xlsx in same folder

print('Dataset shape:', df.shape)
display(df.head())
print("\nColumns:", df.columns.tolist())
```

```python
# The dataset is not in a standard format, each row is a transaction with comma-separated items.
# We need to process it to get a list of transactions, where each transaction is a list of items.
transactions = []
for index, row in df.iterrows():
    # Assuming the single column contains the comma-separated items
    items = str(row.iloc[0]).split(',')
    transactions.append([item.strip() for item in items if item.strip()])

# Display the first few transactions to verify
print("First 5 transactions:")
for i, transaction in enumerate(transactions[:5]):
    print(f"Transaction {i+1}: {transaction}")
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Convert the list of transactions into a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Display the first few rows of the one-hot encoded DataFrame
print("One-hot encoded DataFrame shape:", df_encoded.shape)
display(df_encoded.head())
```

```python
# Apply the Apriori algorithm to find frequent itemsets
# You can adjust the min_support parameter based on your dataset and desired results
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)

print("Frequent Itemsets:")
display(frequent_itemsets.head())

# Generate association rules from the frequent itemsets
# You can adjust the metric and min_threshold parameters
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print("\nAssociation Rules:")
display(rules.head())
```

```python
# Display the rules, sorted by lift
print("Association Rules (sorted by Lift):")
display(rules.sort_values(by='lift', ascending=False).head())

# You can also sort by confidence
# print("\nAssociation Rules (sorted by Confidence):")
# display(rules.sort_values(by='confidence', ascending=False).head())
```

### Interview Questions

1. **What is Lift and Why Important?**  
   Lift measures how much more often the antecedent and consequent appear together than expected if they were independent.  
   \[
   \text{Lift}(A \rightarrow B) = \frac{\text{Confidence}(A \rightarrow B)}{\text{Support}(B)}
   \]
   Lift > 1 means a positive association.

2. **What are Support and Confidence?**  
   - **Support**:  
     Fraction of transactions containing the itemset.  
     \[
     \text{Support}(A) = \frac{\text{Number of transactions with } A}{\text{Total transactions}}
     \]  
   - **Confidence**:  
     Probability that B is purchased when A is purchased.  
     \[
     \text{Confidence}(A \rightarrow B) = \frac{\text{Support}(A \cup B)}{\text{Support}(A)}
     \]

3. **Limitations of Association Rules Mining:**  
   - Generates too many rules  requires good thresholds to prune irrelevant ones.  
   - Only finds pairwise or small combinations  misses global patterns.  
   - Computationally expensive on large datasets.  
   - Doesnt handle continuous variables well.
